### Logistic regression implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
## create the dataset
X , y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=42)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
# model training
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
```

```python
logistic.fit(X_train, y_train)
```

```python
y_pred = logistic.predict(X_test)
print(y_pred)
```

```python
logistic.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
score = accuracy_score(y_test, y_pred)
print(score)
cm = confusion_matrix(y_test, y_pred)
print(cm)
cr = classification_report(y_test, y_pred)
print(cr)
```

### Hperparameter Tuning and Cross validation

```python
model = LogisticRegression()
penalty = ['l1', 'l2', 'elasticnet']
c_values = [100, 10, 1.0, 0.1, 0.01]
solver = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
```

```python
params = dict(penalty=penalty, C=c_values, solver=solver)
```

```python
# GridSearchCV
from sklearn.model_selection import GridSearchCV, StratifiedKFold
cv = StratifiedKFold()
grid = GridSearchCV(estimator=model, param_grid=params, scoring='accuracy', cv=cv, n_jobs=-1)

## estimator is my model, params are my parameter, scoring is finding the accruacy ,
## cv is stratified k fold,
## n_jobs is -1 because we want to use all of our processors
```

```python
grid
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
grid.predict_proba(X_test)
```

### Logistic Regression for Multiclass Classification Problem

```python
## create the dataset
X , y = make_classification(n_samples=1000, n_features=10,n_informative=3, n_classes=3, random_state=42)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression(multi_class='ovr')
logistic.fit(X_train, y_train)
y_pred = logistic.predict(X_test)
```

```python
y_pred
```

```python
score = accuracy_score(y_pred, y_test)
print(score)
print(classification_report(y_pred, y_test))
print(confusion_matrix(y_pred, y_test))
```

```python
model = LogisticRegression(multi_class='ovr')
```

```python
penalty = ['l2', 'l1', 'elasticnet']
c_values = [100, 10, 1, 0.1, 0.01]
solver = ['newton-cg', 'lbfgs', 'sag', 'saga']  # Removed liblinear for multi-class
```

```python
params = dict(penalty=penalty, C=c_values, solver=solver)
```

```python
cv = StratifiedKFold(n_splits=5)  # Added n_splits for clarity
grid = GridSearchCV(estimator=model, param_grid=params, scoring='accuracy', cv=cv, n_jobs=-1)
```

```python
grid.fit(X_train, y_train)
```

```python
print("Best parameters:", grid.best_params_)
print("Best cross-validation score:", grid.best_score_)
```

```python
best_y_pred = grid.predict(X_test)
best_score = accuracy_score(y_test, best_y_pred)
print(f"Best model test accuracy: {best_score}")
```

```python
## more comprihensive
penalty_options = ['l1', 'l2', 'elasticnet']
c_values = [100, 10, 1.0, 0.1, 0.01]
```

```python
# Define compatible solver-penalty combinations
param_grid = [
    {'penalty': ['l2'], 'C': c_values, 'solver': ['newton-cg', 'lbfgs', 'sag', 'saga']},
    {'penalty': ['l1'], 'C': c_values, 'solver': ['liblinear', 'saga']},
    {'penalty': ['elasticnet'], 'C': c_values, 'solver': ['saga'], 'l1_ratio': [0.5]}
]
```

```python
grid_comprehensive = GridSearchCV(estimator=LogisticRegression(multi_class='ovr'),
                                 param_grid=param_grid, scoring='accuracy', cv=cv, n_jobs=-1)
grid_comprehensive.fit(X_train, y_train)
print("Comprehensive search best params:", grid_comprehensive.best_params_)
```
