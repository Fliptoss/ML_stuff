```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/wine_data.csv', header=None, usecols=[0,1,2])
df.columns=['Class label', 'Alcohol', 'Malic acid']
```

```python
df
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malic acid'])
```

```python
color_dict = {1:'blue', 3:'red', 2:'green'}
sns.scatterplot(x=df['Alcohol'], y=df['Malic acid'], hue=df['Class label'], palette=color_dict)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Class label', axis=1), df['Class label'], test_size=0.3, random_state=0)
```

```python
X_train, X_test
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaler.fit(X_train)

X_train_scaler = scaler.transform(X_train)
X_test_scaler = scaler.transform(X_test)
```

```python
X_train_scaler
```

```python
X_train_scaler = pd.DataFrame(X_train_scaler, columns=X_train.columns)
X_test_scaler = pd.DataFrame(X_test_scaler, columns=X_test.columns)
```

```python
np.round(X_train.describe(), 1)
```

```python
np.round(X_train_scaler.describe(), 1)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,5))
ax1.scatter(X_train['Alcohol'], X_train['Malic acid'], c=y_train)
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaler['Alcohol'], X_train_scaler['Malic acid'], c=y_train)
ax2.set_title("After Scaling")
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,5))
ax1.set_title("Before Scaling")
sns.kdeplot(X_train['Alcohol'], ax=ax1)
sns.kdeplot(X_train['Malic acid'], ax=ax1)
ax2.set_title("After Scaling")
sns.kdeplot(X_train_scaler['Alcohol'], ax=ax2)
sns.kdeplot(X_train_scaler['Malic acid'], ax=ax2)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,5))

## alcohol
ax1.set_title("Alcohol before scaling")
sns.kdeplot(X_train['Alcohol'], ax=ax1)

ax1.set_title("Alcohol after scaling")
sns.kdeplot(X_train_scaler['Alcohol'], ax=ax2)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,5))

## alcohol
ax1.set_title("Malic Acid before scaling")
sns.kdeplot(X_train['Malic acid'], ax=ax1)

ax1.set_title("Malic Acid after scaling")
sns.kdeplot(X_train_scaler['Malic acid'], ax=ax2)
```
