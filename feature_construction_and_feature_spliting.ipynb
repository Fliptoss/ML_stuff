```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

import seaborn as sns
```

```python
df = pd.read_csv('/content/train.csv')[['Age', 'Pclass','SibSp','Parch','Survived']]
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
X = df.iloc[:, 0:4]
y = df.iloc[:, -1]
```

```python
X
```

```python
y
```

```python
np.mean(cross_val_score(LogisticRegression(), X, y, scoring='accuracy', cv=20))
```

#### Applying feature construction
feature construction has no sklearn or process. you have to do it manually

```python
X['Family_size'] = X['SibSp'] + X['Parch'] + 1
```

```python
X.head()
```

```python
def myfunc(num):
  if num == 1:
    #alone
    return 0
  elif num > 1 and num <= 4:
    return 1
    ## small family
  else:
    # large family
    return 2
```

```python
myfunc(4)
```

```python
X['Family_type'] = X['Family_size'].apply(myfunc) ## applying for each families
```

```python
X.head()
```

```python
X.drop(columns=['SibSp','Parch','Family_size'],inplace=True)
```

```python
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(), X, y, scoring='accuracy', cv=20))
```

we get a better accuracy than before

### Feature Spliting

```python
df = pd.read_csv('/content/train.csv')
```

```python
df.head()
```

```python
df['Name']
```

## apply seperation names from Mr. etc

```python
df['Title'] = df['Name'].str.split(', ', expand=True)[1].str.split('.', expand=True)[0]
```

```python
df['Name'].str.split(', ', expand=True) ## we choose [1]
```

```python
df['Name'].str.split(', ', expand=True)[1].str.split('.', expand=True)[0]
```

```python
df[['Title','Name']]
```

```python
(df.groupby('Title')['Survived'].mean()).sort_values(ascending=False)  ### likely to have survived
```

```python
df['is_Married'] = 0
df['is_Married'].loc[df['Title'] == 'Mrs'] = 1
```
