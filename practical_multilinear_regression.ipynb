```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/online.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
X = df.drop(['Profit'], axis=1)
y = df['Profit']
```

I should use this pd.get_dummies(X['Area'], drop_first = True). Because as I have already seperated the X and y.

```python
city = pd.get_dummies(X['Area'], drop_first=True).astype(int)
```

```python
city
```

```python
X = pd.concat([X, city], axis=1)
```

```python
X
```

```python
X = X.drop(['Area'], axis=1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
## R2 score
from sklearn.metrics import r2_score
score = r2_score(y_test, y_pred)
```
