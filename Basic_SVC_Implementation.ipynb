```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## create a synthetic data points
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2, n_redundant=0)
```

```python
X
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear')
```

```python
svc.fit(X_train, y_train)
```

```python
# prediction
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
rbf = SVC(kernel='rbf')
rbf.fit(X_train, y_train)
```

```python
y_pred_rbf = rbf.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred_rbf))
print(classification_report(y_test, y_pred_rbf))
```

```python
pol = SVC(kernel='poly')
pol.fit(X_train, y_train)
y_pred_pol = pol.predict(X_test)
print(confusion_matrix(y_test, y_pred_pol))
print(classification_report(y_test, y_pred_pol))
```

```python
sig = SVC(kernel='sigmoid')
sig.fit(X_train, y_train)
y_pred_sig = pol.predict(X_test)
print(confusion_matrix(y_test, y_pred_sig))
print(classification_report(y_test, y_pred_sig))
```

```python
from sklearn.model_selection import GridSearchCV

## defining parameters
param_grid = { 'C': [0.1, 1, 10, 100, 1000], 'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 'kernel': ['rbf']}
```

```python
grid = GridSearchCV(SVC(), param_grid=param_grid, cv=5, refit=True, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred_grid = grid.predict(X_test)
print(classification_report(y_test, y_pred_grid))
print(confusion_matrix(y_test, y_pred_grid))
```
